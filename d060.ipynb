```python
import math

def primeList(nmax):
    '''returns a list of primes up to the the size of nmax'''
    primes = []
    
    n = 2
    
    while n <= nmax:
        if not divisibleByList(n, primes):
            primes.append(n)
        n += 1
        
    return primes
        
            
            
```

```python
def divisibleByList(n, numlist):
    for num in numlist:
        if n%num == 0:
            return True
    return False
```
