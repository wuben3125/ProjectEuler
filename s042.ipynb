# Solution

```python
## put em all together
file = open('p042_words.txt', 'r+')

file_str = ""
for line in file:
    file_str += line.lower()
    
words = file_str.split(',')

file.close
```

```python
# remove puncutation
for idx in range(len(words)):
    words[idx] = words[idx].strip('"')
```

```python
print(words)
```

```python
## set list of char ints
char_ints = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5, 'f':6, 'g':7,
            'h':8, 'i':9, 'j':10, 'k':11, 'l':12, 'm':13,
            'n':14, 'o':15, 'p':16, 'q':17, 'r':18, 's':19,
            't':20, 'u':21, 'v':22, 'w':23, 'x':24, 'y':25, 'z':26}
```

```python
## solution
n = 1
triangle_numbers = [int(n*(n+1)/2)] # instantiate first triangular number
triangle_word_count = 0

for w in words:
    
    word_val = 0 # sum values of chars 
    for char in w:
        word_val += char_ints[char]
    
    
    # makes sure triangle_numbers has enough triangles
    while word_val >= triangle_numbers[-1]:
        # increment largest_n and add triangular numbrs
        n += 1
        
        triangle_numbers.append( int(n*(n+1)/2) )
        
    if word_val in triangle_numbers:
        triangle_word_count += 1
        
print(triangle_word_count)
```
