Find the lowest sum for a set of five primes for which any two primes concatenate to produce another prime.

```python
def update_prime_list(running_prime_list, n):
    """updates running_prime_list to include all primes
    up to and including n"""
    if running_prime_list[-1] < n:
        for potential_prime in range(running_prime_list[-1], n+1):
            prime_check_flag = True

            for prime in running_prime_list:
                if potential_prime % prime == 0:
                    prime_check_flag = False

            if prime_check_flag:
                running_prime_list.append(potential_prime)
```

```python
## first try - using linear prime check

four_primes = [3, 7, 109, 673]

## generate starting primes

running_prime_list = [2]
for potential_prime in range(running_prime_list[-1], four_primes[-1]):
        prime_check_flag = True
        
        for prime in running_prime_list:
            if potential_prime % prime == 0:
                prime_check_flag = False
                
        if prime_check_flag:
            running_prime_list.append(potential_prime)

# instantiation
prime_index = len(running_prime_list)
            
## check for fifth prime
while True:
    
    # if not enough primes, find more
    if prime_index >= len(running_prime_list):
        # overkill for safety
        update_prime_list(running_prime_list, 2*running_prime_list[-1])
    
    fifth_prime = running_prime_list[prime_index]
    
    # test if new concatenations are also prime
    successful_set_flag = True
    
    for i in range(4):
        concat_prime = concat_nums(four_primes[i], fifth_prime)
        if not check_prime(running_prime_list, concat_prime):
            successful_set_flag = False
            break
        
        concat_prime = concat_nums(fifth_prime, four_primes[i])
        if not check_prime(running_prime_list, concat_prime):
            successful_set_flag = False
            break
    
    # check
    if successful_set_flag:
        five_primes = four_primes
        five_primes.append(fifth_prime)
        
        print(f"These are the five primes: {five_primes}")
        print(f"The upper bound sum is: {sum(five_primes)}")
        break # done
        
    # update prime_index
    prime_index += 1
```
