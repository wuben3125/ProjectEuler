```python
### 1.
def word_to_value(word, char_prime_dict):
#     value = 0 # mistake
    value = 1
    for char in word:
#         value += char_prime_dict[char] # mistake
        value *= char_prime_dict[char]
        
    return value
```

```python
## testing
word_to_value('abc', {'a':2, 'b':3, 'c':5})
```

```python
### 2.
def squares_up_to_length_n(n):
#     upper_bound = 10**(n+1) # mistake
    upper_bound = 10**n # if n is 0, [0, 1] is returned
    square_list = []
    
    base = 0
    while True:
        square = base**2
        
        if square < upper_bound:
            square_list.append(square)
        else:
            break # end of list
            
        base += 1 # forgot this
            
    return square_list
```

```python
## testing
squares_up_to_length_n(2)
```

```python
### 3.
def binary_search(ascending_list, target): 
    low = 0
    high = len(ascending_list) -1
    
    while(low < high):
#         mid = int((low+1 + high)/2)
        mid = int((low+high)/2)
        
        if ascending_list[mid] == target:
            return True
        elif ascending_list[mid] < target:
#             low = mid
            low = mid + 1
        else:
            high = mid
            
    return False
```

```python
## testing
print(binary_search([1,2,3,4,5], 2))
print(binary_search([1,2,3,4,5], -1))
```

```python
### 4. 
def digit_permutations_of_length_n(n):
    import itertools
    
    return [perm for perm in itertools.permutations(range(10), n)]
    
```

```python
## testing
digit_permutations_of_length_n(2)
```

```python
### 4.5
def word_to_letter_index_dict(word):
    letter_index_dict = {}
    
    for i in range(len(word)):
        char = word[i]
        
        if char not in letter_index_dict.keys():
            letter_index_dict[char] = [i]
        else:
            letter_index_dict[char].append(i)
            
    return letter_index_dict
```

```python
## testing
print(word_to_letter_index_dict('bobby'))
print(word_to_letter_index_dict('ybbob'))
```

```python
### 5. 
def digit_perm_to_int(letter_index_dict, digit_perm, n): 
    # sums powers of ten according to each key
        # order doesn't matter
    output = 0 
    
    digit_index = 0 # indices of digit_perm
    
    for key in letter_index_dict.keys():
#     for key in sorted(letter_index_dict.keys()): # for consistent ordering
        for letter_index in letter_index_dict[key]:
            
            ## ___(Start)___
            # room to weed out first digit cases
                # 2, 3, 7, 8
            if letter_index == n-1:
                if digit_perm[digit_index] in [2,3,7,8]:
                    return -1 # the output won't be square
                
            ## ___(End)____
            
            if letter_index == 0: # leading letter
                if digit_perm[digit_index] == 0: # special case
                    return -1 # can't have leading 0
                else: # normal case
                    output += digit_perm[digit_index] * 10**(n-1 - letter_index)
            
            else:
                output += digit_perm[digit_index] * 10**(n-1 - letter_index)
    
        digit_index += 1 # move to next digit in perm
    
    return output
    
# note: this wasn't well thought-out...
# also, earlier I was confinishing confining myself too thinly
```

```python
### get a list of anagram pairs

## get words
file = open('p098_words.txt')
words = [word.strip('"') for word in file.readline().split(',')]

# # testing
# print(words)

value_word_dict = {}

letter_prime_dict = {
    'A': 2, 'B': 3, 'C': 5, 'D': 7, 'E': 11, 'F': 13,
    'G': 17, 'H': 19, 'I': 23, 'J': 29, 'K': 31, 'L': 37,
    'M': 41, 'N': 43, 'O': 47, 'P': 53, 'Q': 59, 'R': 61,
    'S': 67, 'T': 71, 'U': 73, 'W': 79, 'X': 83, 'Y': 87,
    'Z': 89, 'V': 91
}

## sift out anagram words
for word in words:
    value = word_to_value(word, letter_prime_dict)
    
    if value not in value_word_dict.keys():
        value_word_dict[value] = [word] # create new entry
    else:
        value_word_dict[value].append(word)

# get pairs        
anagram_list = []
for value in value_word_dict.keys():
    if len(value_word_dict[value]) >= 2:
        anagram_list.append(value_word_dict[value])
        
## testing
print(anagram_list)
```
