# Longest Collatz Sequence
source: https://projecteuler.net/problem=14

## Problem Statement
A Collatz sequence is generated by following the steps:
* n --> n/2 (if n is even)
* n --> 3n + 1 (if n is odd)

It's an unproven conjecture that all Collatz sequences terminate in 1.

Anyways, what's the longest Collatz sequences whose starting number is under 1,000,000?

(Implicitly assumed that starting numbers are positive)

## Planning
Pretty much brute force

Make efficient function to find length of Collatz Sequence

Parse through all numbers from 10^6-1 to 1, compare largest each time

## [work]

```python
def collatz_length(n):
    """
    n: int
    
    returns length of Collatz sequence starting from n
    
    doesn't keep track of actual sequence
    """
    
    length = 1
    
    while n > 1:
        if n%2==0: n//=2
        else: n = 3*n+1
        
        length += 1
        
    return length
```

```python
# testing
collatz_length(13)

# works, checked with Project Euler page example
```

## Solution

```python
import timeit
```

```python
## testing range

for i in range(10, 0, -1): print(i)
```

```python
largest_length = 1
starting_num = 0

start = timeit.default_timer()

for i in range(10**6-1, 0, -1):
    if collatz_length(i) > largest_length:
        starting_num = i
        largest_length = collatz_length(starting_num)
        
print(f"The sequence beginning with {starting_num} has {largest_length} elements")

end = timeit.default_timer()
print(f"Time elapsed: {end-start}")
```
