```python
# function to find first four decimal digits

def get_first_four_digits(n):
    digits = int(10000*(1/n))/10000
    
    digits_list = list(str(digits)[2:])

    digits_list = [int(n) for n in digits_list]
    
    return digits_list
```

```python
print(get_first_four_digits(243))
```

```python
# function to parse through division digits, stop when first four digits are repeated, return length of sequence
def sequence_length(n):
#     first_four_digits = get_first_four_digits(n) # doesn't work; need to find the first four potential repeats
    
    current_four_digits = []
    
    running_dividend = 1
    
    count = 0
    
#     while running_dividend%n!=0 and current_four_digits!=first_four_digits:
    
    past_first_instance = False
    
    while running_dividend%n!= 0:
        
        if(len(current_four_digits) == 4): current_four_digits.pop(0)

        
        if running_dividend < n:
            running_dividend *= 10
            current_four_digits.append(0)
            
        else:
            current_four_digits.append(running_dividend//n)
            running_dividend %= n
            running_dividend *=10
        
        if current_four_digits == first_four_digits:
            if not past_first_instance: 
                past_first_instance = True
                            
            else:
                break
    
        count += 1
        
        print(first_four_digits)
        print(current_four_digits)
        
    return count
```

## Solution

```python
import timeit

largest_length = 0
largest_length_d = 0

start = timeit.default_timer()

for i in range(1, 1000+1):
    
    temp = sequence_length(i)
    
    if temp > largest_length:
        largest_length = temp # avoids redundant calculations
        largest_length_d = i
        
        print(largest_length)
        
print(largest_length_d)

end = timeit.default_timer()
```

```python
print('time elapsed:', (end-start))
print('number of elements in sequence:', (largest_length-4))
```
