# Triangular, pentagonal, and hexagonal
Source: https://projecteuler.net/problem=45

# Problem Statement:
Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

Triangle	 	Tn=n(n+1)/2	 	1, 3, 6, 10, 15, ...

Pentagonal	 	Pn=n(3n−1)/2	 	1, 5, 12, 22, 35, ...

Hexagonal	 	Hn=n(2n−1)	 	1, 6, 15, 28, 45, ...

It can be verified that T285 = P165 = H143 = 40755.

Find the next triangle number that is also pentagonal and hexagonal.

# Planning
Iterate over triangular numbers starting with T_285, and compare to a list of pentagonal and hexagonal numbers

```python
# list tinkering
[1,2,3] + [4,5]
```

```python
import timeit
```

```python
n_tri = 285
n_pent = 165
n_hex = 143

tri_list = [int(n_tri*(n_tri+1)/2)]
pent_list = [int(n_pent*(3*n_pent-1)/2)]
hex_list = [int(n_hex*(2*n_hex-1))]

start = timeit.default_timer()

while True:
    
    n_tri += 1
    
    tri_list.append( int(n_tri*(n_tri+1)/2) )
    
    ## update pent_list
    while tri_list[-1] > pent_list[-1]:
        n_pent += 1
        pent_list.append( int(n_pent*(3*n_pent-1)/2) ) 
    
    ## update hex_list
    while tri_list[-1] > hex_list[-1]:
        n_hex += 1
        hex_list.append( n_hex * (2*n_hex - 1))
    
    if tri_list[-1] in pent_list and tri_list[-1] in hex_list:
        print(f"{tri_list[-1]}, for n_tri={n_tri}, n_pent={n_pent}, n_hex={n_hex}")
        end = timeit.default_timer()
        
        print(f"time elapsed: {end-start}")
        print()
```
