```python
# Smallest multiple:

# 2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

# What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?
```

```python
# Planning:

# I could find an arbitrary crunch method, but I'll give myself 10 minutes to think of an efficient solution this time

# What I did is, starting from the top, parsed through all integers under n
# if the integer divides into n, we ignore it
# but if it isn't, we multiply the product (default value n) by the uncommon factor
# repeat until 2 is reached, since 1 is trivial

# I could try this with recursive code, see if it's faster?
```

```python
# Code:

import math
```

```python
def smallestMultipleFactorsUpToN(n):
    product = n
    
    i = n-1 # counter
    while i > 1:
        if (product % i) != 0:
            product *= int(i/(math.gcd(product,i)))
        i -= 1
    return product
```

```python
# Solution:

smallestMultipleFactorsUpToN(20)
```

```python
smallestMultipleFactorsUpToN(1000)
```

```python
smallestMultipleFactorsUpToN(100000)
# took like, 15 seconds
```

```python
def recursiveSmallestMultipleFactorsUpToN(n):
    return recursiveHelperSmallestMultipleFactorsUpToN(n-1, n)
```

```python
def recursiveHelperSmallestMultipleFactorsUpToN(i, product):
    if i <= 1:
        return product
    else:
        if (product % i) != 0:
            product *= int(i/math.gcd(i, product))
        return recursiveHelperSmallestMultipleFactorsUpToN(i-1, product)
    
```
