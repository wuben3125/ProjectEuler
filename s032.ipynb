```python
def check_identity(multiplicand, multiplier, product):
    return multiplicand*multiplier == product
```

```python
# timeit stuff added as a special touch
import timeit

start = timeit.default_timer()

pandigital_products = set()

for perm in itertools.permutations(range(1, 9+1)):
    ## digit split of 1, 4, 4 (same as 4, 1, 4)
#     product = number_from_tuple(perm, 5, 8)
    multiplicand = number_from_tuple(perm, 0, 0)
    multiplier = number_from_tuple(perm, 1, 4)
    product = number_from_tuple(perm, 5, 8)

    
    if check_identity(multiplicand, multiplier, product):
#         set.add(product)
#         pandigital_product.add(product)
        pandigital_products.add(product)


    ## digit split of 2, 3, 4 (same as 3, 2, 4)

    multiplicand = number_from_tuple(perm, 0, 1)
    multiplier = number_from_tuple(perm, 2, 4)
    product = number_from_tuple(perm, 5, 8)
    
    if check_identity(multiplicand, multiplier, product):
#         set.add(product)
#         pandigital_product.add(product)
        pandigital_products.add(product)
        
        
end = timeit.default_timer()
```
