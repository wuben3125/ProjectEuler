```python
def primes_under_n(n):
    if n < 2:
        return []
    else:
        prime_list = [2]
        
        for num in range(3, n+1):
            prime_flag = True
            for prime in prime_list:
                if num%prime == 0:
                    prime_flag = False
                    
            if prime_flag:
                prime_list.append(num)
                
        return prime_list
```

```python
## testing
primes_under_n(25)
```

```python
def prime_factors(n):
#     prime_list = primes_under_n(n)
    prime_list = primes_under_n(int(n/2))
    factors = []
    
    for prime in prime_list:
        if n%prime == 0:
            factors.append(prime)
            
    return factors
```

```python
## testing
prime_factors(3*5*7*11*13)
```

```python
def all_prime_factors(n):
    factors = prime_factors(n)
    all_factors = []
    
    for prime in factors:
        temp_n = n
        while temp_n%prime == 0:
            all_factors.append(prime)
            temp_n /= prime
            
    return all_factors
```

```python
## testing
all_prime_factors(5**5)
```

```python
def solution(n):
    # instantiate variables
    running_prime_list = [2]
    
    first_number = 1
    prime_factor_counts = [0 for i in range(n)]
    
        
    # instantiation
    while True:
        
        all_n_satisfied_flag = True
        
        # checks n consecutive numbers at a time
        for i in range(n):
            current_num = first_number + i
            
            # update running prime list if needed
            if running_prime_list[-1] < current_num/2:
                for potential_prime in range(running_prime_list[-1] + 1, int(current_num/2)):
                    prime_flag = True
                    for prime in running_prime_list:
                        if potential_prime % prime == 0:
                            prime_flag = False
                    
                    if prime_flag:
                        running_prime_list.append(potential_prime)
            
#             print(running_prime_list)
            
            # count number of prime factors
            prime_count = 0
            
            for prime in running_prime_list:
                if current_num % prime == 0:
                    prime_count += 1
                
                if prime_count >= n:
                    break # no need to count beyond
                    
            if prime_count < n:
                # the sequence doesn't satisfy the solution condition
                
                # shift up first_number
                first_number = current_num + 1
                
                all_n_satisfied_flag = False
                
                break
            
        if all_n_satisfied_flag:
            return first_number
    
    
```
